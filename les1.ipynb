```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('retail_data_set/retail_sales.csv')
df.head()
```

```python
plt.scatter(df['Product Category'], df['Sales Revenue (USD)'])
plt.xlabel('Store Location')
plt.ylabel('Units Sold')
plt.title('Scatterplot van Product Catergory tov Sales Revenue')

plt.show()
```
